Disbursal Date Range

```python
#Convert [DISBURSAL_DATE] column into datetime using pd.to_datetime()

loan_df['DISBURSAL_DATE'] = pd.to_datetime(loan_df['DISBURSAL_DATE'], format = 'mixed')

print('Max disbursal date: ', loan_df['DISBURSAL_DATE'].max())
print('Minimum disbursal date: ', loan_df['DISBURSAL_DATE'].min())
print('time_span: ', loan_df['DISBURSAL_DATE'].max() - loan_df['DISBURSAL_DATE'].min())
```
